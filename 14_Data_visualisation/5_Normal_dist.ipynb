```python
import numpy as np
import matplotlib.pyplot as plt

# To get the number and it's frequency 
from collections import Counter
```

### Getting a sample of Standard normal dist
Return a sample (or samples) from the "standard normal" distribution.

```python
xsn = np.random.randn(100)
xsn = np.sort(xsn,kind = "mergesort")
print(type(xsn))

# To get the number and it's frequency 
# important no need of converting it to dict as heave pre- defined functions .keys() and .values()
num_freq = Counter(xsn)
```

```python
num_freq
```

### <center>Converting standard normal distribution to normal distribution<center>


```python
# Let us assume that standard deviation is 5
sigma = 5

# Let us assume that mean of sample is 60
u = 60

X = xsn*sigma + u
print(X)
```
