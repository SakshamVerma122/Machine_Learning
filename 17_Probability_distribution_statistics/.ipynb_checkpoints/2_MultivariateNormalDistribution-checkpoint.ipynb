### Multi variate Normal Distribution

```python
# Importing the libraries
import matplotlib.pyplot as plt
import numpy as np

# Bivariate
mean = np.array([0,0])
# Here there is no corr b/w x and y
cov = np.array([[1,0],[0,1]])

dist = np.random.multivariate_normal(mean,cov,500)
print(dist.shape)

# Seperating variables
X = dist[:,0]
Y = dist[:,1]

# Plotting scatter plot
plt.scatter(X,Y)
plt.show()
```

### Varing the spread of the data along x and y axis via varing the sigma(xx) and sigma(yy)

```python
# Moved the mean to (5,6)
mean2 = np.array([5,6])
# Here there is no corr b/w x and y
# increased the spread of x and y
# if you change only one then spread will increase only wrt that axis
cov2 = np.array([[2,0],[0,2]])


dist2 = np.random.multivariate_normal(mean2,cov2,500)
print(dist.shape)

# Seperating variables
X2 = dist2[:,0]
Y2 = dist2[:,1]

# Plotting scatter plot
plt.scatter(X2,Y2)
# Plotting the first one too as to give you a comparison
plt.scatter(X,Y)
plt.show()
```
