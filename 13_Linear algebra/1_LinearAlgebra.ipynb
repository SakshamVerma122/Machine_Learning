```python
# uint8 means -->unsigned integer of 8 bits 
# It's perfect memory allocation for images as 
T = np.zeros([5,5,3],dtype = np.uint8)
# T = np.ones(((5,5,3),dtype="uint8")
print(T)

import matplotlib.pyplot as plt

# The image is 2D as the pixel value is combined to form colour
plt.imshow(T)
plt.show()
```

```python
# Basically we converted every Blue in RGB to 255
T[:,:,2] = 255
plt.imshow(T)
plt.show()
T
```

```python
T[:,:,(0,2)] = 255
plt.imshow(T)
plt.show()
T
```

```python
# Reinitialising to 0
T[:,:,:] = 0

# Changing every 2nd to 255
T[:,:,::2] = 255
plt.imshow(T)
plt.show()
T
```

### Appending a row
A copy of arr with values appended to axis. Note that append does not occur in-place: a new array is allocated and filled. If axis is None, out is a flattened array.

```python
C = np.array(([1,2,3,3,4,5]))
print(np.append(C,[1,2]))
print(np.append(C,1))
```

```python
# Here array is flattened as axis = None
np.append(A,[1,2])
```

```python
# Here error occured as both don't have same dimension
np.append(A,[1,2],axis = 0)
```

```python
# Appending a row array
np.append(A,[[1,2]],axis = 0)
```

```python
# Appending a column array
np.append(A,[[1],[2]],axis = 1)
```

```python
np.append(A,[[1,2]])
```
