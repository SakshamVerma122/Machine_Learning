```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
browser = webdriver.Chrome()
# For different Web browsers version abd browsers you need to install different selenium versions
# browser = webdriver.Chrome("Path to the chrome")
```

```python
browser.get("https://www.codechef.com/login?destination=/")
```

##### Finding elements via id

```python
username_element = browser.find_element(By.ID,"edit-name")

# Finding elements via class
# browser.find_elements_by_class_name("m-login-button")
```

```python
username_element
```

#####  USING X-PATH to send user name

```python
username_element = browser.find_element_by_xpath('/html/body/section/div[2]/div/div[3]/div[2]/form/div/div[1]/div/div/input')
```

##### pushing the text into username

```python
username_element.send_keys("sakshamv001")
```

```python
password_element = browser.find_element_by_id("edit-pass")
```

```python
userpass_element = browser.find_element_by_xpath('/html/body/section/div[2]/div/div[3]/div[2]/form/div/div[2]/div/div[2]/input')
```

```python
from getpass import getpass
```

```python
userpass_element.send_keys(getpass("Enter password:"))
```
