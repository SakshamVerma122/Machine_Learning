```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2
print(X,Y)
```

### Given a Function f(x),we want to find the value of x that minimises f(x).

```python
plt.style.use("seaborn")
plt.plot(X,Y)
plt.xlabel("x")
plt.ylabel("f(x)")
```

```python
x = 0

#Step size
lr = 0.1
error = []

# take 50 steps in the downhill direction
for i in range(50):
    
    # grad is differntiation of f(x)
    grad = 2*(x-5)
    
    y = (x-5)**2
    error.append(y)
    plt.plot(x,y,"o")
    
    # we subtracted it with lr*grad as it will always try to move toward the minima as slope values sign will always lead it to there
    x = x-lr*grad
plt.plot(Y)    
```

### Plot the values of error

```python
# Here we can see that as plot is approaching to the minima errors value 
# is decreasing and after a point it is about constant
plt.plot(error)
```
